# Plotting vectors

```python
plt.quiver(0,0,3,4)     initial=(0,0) going towards=(3,4)
plt.show()
```

```python
plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale = 1,color='red')
plt.quiver(0,0,-3,4,scale_units='xy',angles='xy',scale = 1,color='green')

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
def plot_vectors(vecs):
    col = ['b','g','r','y']
    i = 0
    for vec in vecs:
        plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units='xy',angles='xy',scale = 1,color=col[i%len(col)])
        i+=1
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()
```

```python
plot_vectors([(0,0,3,4), (0,0,-3,4),(0,0,1,5),(0,0,5,1)])
```

# Addition and Substraction

```python
vecs = [(0,0,3,4), (0,0,-3,4),(0,0,1,5),(0,0,5,1)]
```

```python
plot_vectors([vecs[0],vecs[3]])
```

```python
print(vecs[0]+vecs[1])
```

```python
vecs = [np.asarray([0,0,3,4]), np.asarray([0,0,-3,4]),np.asarray([0,0,1,5]),np.asarray([0,0,5,1])]
```

```python
print(vecs[0]+vecs[1])
```

```python
plot_vectors(vecs)
```

```python
plot_vectors([vecs[0],vecs[3],vecs[0]+vecs[3]])
```
